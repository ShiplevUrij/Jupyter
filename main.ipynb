```python
import math
from sympy import *
from sympy.plotting import plot
init_printing()

x = Symbol(' x ')

f = -12*x**4*sin(cos(x)) - 18*x**3+5*x**2 + 10*x - 30
f
```

```python
# Определить корни
solveset(f, x)
```

```python
# Найти интервалы, на которых функция возрастает
# Найти интервалы, на которых функция убывает
print(solveset(f > 0,x))
print(solveset(f < 0,x))
```

```python
fun=plot(f, show=False)
fun.show()
```
